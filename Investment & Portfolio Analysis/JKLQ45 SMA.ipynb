```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook

#opening the file and making the date column as index
data = pd.read_csv("JKLQ45.csv", index_col=0, parse_dates=True)
data['Close']
```

```python
#calculating SMA 5 days
data['SMA5'] = data['Close'].rolling(5).mean()
data.tail()
```

```python
#calculating SMA 30 days
data['SMA30'] = data['Close'].rolling(30).mean()
data.tail()
```

```python
#calculating SMA 50 days
data['SMA50'] = data['Close'].rolling(50).mean()
data.tail()
```

```python
top_plt = plt.subplot2grid((5,4), (0, 0), rowspan=3, colspan=4)
top_plt.plot(data['Close'].loc['2021-01-01':].index, data['Close'].loc['2021-01-01':], label='Closing Price')
top_plt.plot(data['Close'].loc['2021-01-01':].index, data['SMA5'].loc['2021-01-01':], label='SMA5')
top_plt.plot(data['Close'].loc['2021-01-01':].index, data['SMA30'].loc['2021-01-01':], label='SMA30')
top_plt.plot(data['Close'].loc['2021-01-01':].index, data['SMA50'].loc['2021-01-01':], label='SMA50')
plt.xlabel('Time Period')
plt.ylabel('Historical Prices')
plt.title('JKLQ45 Index')
plt.legend()
bottom_plt = plt.subplot2grid((5,4), (3,0), rowspan=1, colspan=4)
bottom_plt.bar(data['Volume'].loc['2021-01-01':].index, data['Volume'].loc['2021-01-01':])
plt.title('\nTrading Volume', y=-0.60)
plt.gcf().set_size_inches(10,8)
plt.tight_layout()
```
