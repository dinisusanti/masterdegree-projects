```python
#calculating Close data volatility with average (490/2) trading days per year
volatility = data['Log Returns'].std()*245**.5
volatility
```

```python
#changing the value of volatility into percentage
str_vol = str(round(volatility, 4)*100)
str_vol
```

```python
#calculating MACD
exp1 = data['Close'].ewm(span=12, adjust=False).mean()
exp2 = data['Close'].ewm(span=26, adjust=False).mean()
data['MACD'] = exp1-exp2
data['Signal Line'] = data['MACD'].ewm(span=9, adjust=False).mean()
data.tail()
```

```python
#plotting MACD and Signal Line
fig, ax = plt.subplots()
data[['MACD', 'Signal Line']].loc['2022-01-01':].plot(ax=ax)
data['Close'].loc['2022-01-01':].plot(ax=ax, alpha=0.25, secondary_y=True)
```

```python
#calculating stochastic oscillator
high14 = data['High'].rolling(14).max()
low14 = data['Low'].rolling(14).min()
data['%K'] = (data['Close'] - low14)*100/(high14-low14)
data['%D'] = data['%K'].rolling(3).mean()
data.tail()
```

```python
#plotting the stochastic oscillator
fig, ax = plt.subplots()
data[['%K', '%D']].loc['2022-01-01':].plot(ax=ax)
ax.axhline(80, c='r', alpha=0.3)
ax.axhline(20, c='r', alpha=0.3)
data['Close'].loc['2022-01-01':].plot(ax=ax, alpha=0.3, secondary_y=True)
```

```python
#calculating RSI
data['delta'] = delta = data['Close'].diff()
data['up'] = up = delta.clip(lower=0)
data['down'] = down = -1*delta.clip(upper=0)

ema_up = up.ewm(com=14, adjust=False).mean()
ema_down = down.ewm(com=14, adjust=False).mean()

rs = ema_up/ema_down
data['RSI'] = 100 - (100/(1 + rs))
data.tail()
```

```python
#plotting RSI
fig, ax = plt.subplots()
data['RSI'].loc['2022-01-01':].plot(ax=ax)
ax.axhline(30, linestyle='--', c='g', alpha=0.3)
ax.axhline(50, linestyle='--', c='orange', alpha=0.3)
ax.axhline(70, linestyle='--', c='r', alpha=0.3)
```
